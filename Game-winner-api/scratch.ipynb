```python
from datetime import datetime, timedelta
from nba_api.stats.endpoints import scoreboard, scoreboardv2
import pandas as pd

# get yesterday's date
yesterday = datetime.now() - timedelta(days=1)
yesterday_str = yesterday.strftime('%m/%d/%Y')

# get scoreboard for yesterday's date
games = scoreboardv2.ScoreboardV2(day_offset=0, game_date=yesterday_str).line_score.get_data_frame()

games.head()
# loop through each game and print the winner
# for index, game in games.iterrows():
#     print(f"{game['VISITOR_TEAM_NAME']} {game['VISITOR_TEAM_SCORE']} - {game['HOME_TEAM_SCORE']} {game['HOME_TEAM_NAME']}")
#     if game['VISITOR_TEAM_SCORE'] > game['HOME_TEAM_SCORE']:
#         print(f"Winner: {game['VISITOR_TEAM_NAME']}")
#     else:
#         print(f"Winner: {game['HOME_TEAM_NAME']}")
```

```python
def get_winner(group):
    points = group['PTS'].idxmax()
    winner = group.loc[points, ['GAME_DATE_EST', 'GAME_SEQUENCE', 'TEAM_ID']]
    return winner

winners = games.groupby('GAME_ID').apply(get_winner)
print(winners)
```
